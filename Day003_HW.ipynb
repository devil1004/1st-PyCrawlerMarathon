### 比較一下範例檔案中的「File I/O」與「xmltodict」讀出來的內容有什麼差異


```python
'''
Your Code
'''
import os
import xmltodict



fh = open(r"D:\1st-PyCM\data\Data\64_72hr_CH.xml", "r",encoding='utf-8')
xml = fh.read()
fh.close()

d = dict(xmltodict.parse(xml))
```

### 根據範例檔案的結果：

1. 請問高雄市有多少地區有溫度資料？
2. 請取出每一個地區所記錄的第一個時間點跟溫度
3. 請取出第一個地區所記錄的每一個時間點跟溫度

```python
#1. 請問高雄市有多少地區有溫度資料？

ka = d['cwbopendata']['dataset']['locations']['location']

print(f'高雄市有{len(ka)}個地區有資料')
for lo in range(len(ka)):
    print(ka[lo]['locationName'])
```

```python
# 2. 請取出每一個地區所記錄的第一個時間點跟溫度

for lo in range(len(ka)):
    lo_time = ka[lo]['weatherElement'][0]['time'][0]
    print(f"地區: {ka[lo]['locationName']}   時間:{lo_time['dataTime']}   溫度:{lo_time['elementValue']['value']}")
```

```python
# 3. 請取出第一個地區所記錄的每一個時間點跟溫度

temp = ka[0]['weatherElement'][0]['time']

for time in range(len(temp)):
    
    print(f"時間:{temp[time]['dataTime']} 溫度:{temp[time]['elementValue']['value']}")
```
