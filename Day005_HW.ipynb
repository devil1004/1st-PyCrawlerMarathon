```python
# 1. é€™å€‹ API ä¸€æ¬¡æœƒå›å‚³å¹¾ç­†è³‡æ–™ï¼Ÿæ¯ä¸€ç­†è³‡æ–™åŒ…å«å“ªäº›æ¬„ä½ï¼Ÿ
import requests,json

r = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
re = json.loads(r.text)

print(f'æ¯æ¬¡å›å‚³ {len(re)} ç­†è³‡æ–™')

print(f'æ¯ç­†è³‡æ–™åŒ…å« {list(re[0].keys())}')
```

```python
# 2. å–å‡ºæ¯ä¸€ç­†è³‡æ–™çš„ã€Œæ¨™é¡Œã€ã€ã€Œè²¼æ–‡æ™‚é–“ã€ã€ã€Œç•™è¨€äººæ•¸ã€ã€ã€ŒæŒ‰è®šäººæ•¸ã€

for d in re:
    print(f"æ¨™é¡Œ: {d['title']}")
    print(f"è²¼æ–‡æ™‚é–“: {d['createdAt']}")
    print(f"ç•™è¨€äººæ•¸: {d['commentCount']}")
    print(f"æŒ‰è®šäººæ•¸: {d['likeCount']}\n")
```

```python
# 3. è¨ˆç®—ç†±é–€/éç†±é–€æ–‡ç« çš„ã€Œå¹³å‡ç•™è¨€äººæ•¸ã€èˆ‡ã€Œå¹³å‡æŒ‰è®šäººæ•¸ã€

r_np = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=false')
re_np = json.loads(r_np.text)

count = [0,0,0,0]

for d in re:
    count[0] += d['commentCount']
    count[1] += d['likeCount']

for d in re_np:
    count[2] += d['commentCount']
    count[3] += d['likeCount']
    
print("ç†±é–€æ–‡ç« :\n\
    å¹³å‡ç•™è¨€äººæ•¸: {:.2f}\
    å¹³å‡æŒ‰è®šäººæ•¸: {:.2f}".format(count[0]/30,count[1]/30))

print("æœ€æ–°æ–‡ç« :\n\
    å¹³å‡ç•™è¨€äººæ•¸: {:.2f}\
    å¹³å‡æŒ‰è®šäººæ•¸: {:.2f}".format(count[2]/30,count[3]/30))
```
